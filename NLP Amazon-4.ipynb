```python
import nltk
from nltk.tokenize import wordpunct_tokenize
from nltk.corpus import stopwords 
import string

reviews_text = data['reviews.text']
stop_words = stopwords.words('english')
```

```python
def prepare_text(text):
    tokens = [word for word in nltk.wordpunct_tokenize(text.lower()) if word not in string.punctuation and word not in stop_words]
    text_without_punctuation = ' '.join(tokens)
    for i in string.punctuation:
        text_without_punctuation = text_without_punctuation.replace(i, '')
    prepared_text = text_without_punctuation.rstrip().lstrip()
    return prepared_text
```

```python
data['reviews.preparedText'] = data['reviews.text'].apply(prepare_text)
data['reviews.title'] = data['reviews.title'].apply(prepare_text)
```

```python
from sklearn.model_selection import train_test_split

data_train, data_test = train_test_split(data, test_size=0.35, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

cat_cols = ['id', 'brand', 'categories', 'colors', 'manufacturer', 'name', 'reviews.doRecommend', 'sentiment']
num_cols = ['reviews.numHelpful', 'reviews.rating', 'price']
le = LabelEncoder()
ss = StandardScaler()

for col in cat_cols:
    data_train[col] = le.fit_transform(data_train[col])
    data_test[col] = le.transform(data_test[col])
    
for col in num_cols:
    data_train[[col]] = ss.fit_transform(data_train[[col]])
    data_test[[col]] = ss.transform(data_test[[col]])
```

```python
import gensim
from gensim.models.ldamodel import LdaModel
from gensim import corpora

# Creating the term dictionary of our courpus, where every unique term is assigned an index.
doc_prepared_train = list(doc.split() for doc in data_train['reviews.preparedText'])
doc_prepared_test = list(doc.split() for doc in data_test['reviews.preparedText'])

dictionary_train = corpora.Dictionary(doc_prepared_train)
dictionary_test = corpora.Dictionary(doc_prepared_test)

# Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above.
doc_term_matrix_train = [dictionary_train.doc2bow(doc) for doc in doc_prepared_train]
doc_term_matrix_test = [dictionary_test.doc2bow(doc) for doc in doc_prepared_test]

lda_train = LdaModel(doc_term_matrix_train, num_topics=5, id2word = dictionary_train, passes=50)
lda_test = LdaModel(doc_term_matrix_test, num_topics=5, id2word = dictionary_test, passes=50)
```
