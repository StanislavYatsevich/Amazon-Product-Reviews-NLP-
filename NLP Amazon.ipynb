```python
def rating_to_recommend(rating):
    return 'true' if rating > 3 else 'false'

ratings_to_recommend = {i : rating_to_recommend(i) for i in data['reviews.rating'].unique()}
data['reviews.doRecommend'].fillna(data['reviews.rating'].map(ratings_to_recommend), inplace=True)
```

```python
def rating_to_sentiment(rating):
    if rating > 3.5:
        return 'positive'
    if rating >= 2.5 and rating <= 3.5:
        return 'neutral'
    else:
        return 'negative'
rating_to_sentiment = {i : rating_to_sentiment(i) for i in data['reviews.rating'].unique()}
data['sentiment'] = data['reviews.rating'].map(rating_to_sentiment)
```

```python
import nltk
from nltk.tokenize import wordpunct_tokenize
from nltk.corpus import stopwords 
import string

reviews_text = data['reviews.text']
stop_words = stopwords.words('english')
```

```python
def prepare_text(text):
    tokens = [word for word in nltk.wordpunct_tokenize(text.lower()) if word not in string.punctuation and word not in stop_words]
    text_without_punctuation = ' '.join(tokens)
    for i in string.punctuation:
        text_without_punctuation = text_without_punctuation.replace(i, '')
    prepared_text = text_without_punctuation.rstrip().lstrip()
    return prepared_text
```

```python
data['reviews.preparedText'] = data['reviews.text'].apply(prepare_text)
```

```python
from wordcloud import WordCloud

text = ' '.join(data['reviews.preparedText'])
wc = WordCloud().generate(text)

fig = px.imshow(wc, title="Results of the WordCloud model", text_auto=True)
fig.show()
```
